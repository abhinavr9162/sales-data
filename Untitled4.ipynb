```python
import pandas as pd
df = pd.read_csv("/content/sample_data/sales_data_sample.csv",encoding='unicode_escape')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

**handilling missing value**

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df['SALES'].fillna(df['SALES'].mean(), inplace=False)
```

handling duplicate vaue

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

```python
df['ORDERDATE'] = pd.to_datetime(df['ORDERDATE'])
df['ORDERDATE']
```

```python
df = df.rename(columns={
    'ORDERNUMBER': 'order_number',
    'QUANTITYORDERED': 'quantity_ordered',
    'PRICEEACH': 'price_each',
    'ORDERLINENUMBER': 'order_line_number',
    'SALES': 'sales',
    'ORDERDATE': 'order_date',
    'STATUS': 'status',
    'QTR_ID': 'quarter_id',
    'MONTH_ID': 'month_id',
    'YEAR_ID': 'year_id',
    'PRODUCTLINE': 'product_line',
    'MSRP': 'msrp',
    'PRODUCTCODE': 'product_code',
    'CUSTOMERNAME': 'customer_name',
    'PHONE': 'phone',
    'ADDRESSLINE1': 'address_line1',
    'ADDRESSLINE2': 'address_line2',
    'CITY': 'city',
    'STATE': 'state',
    'POSTALCODE': 'postal_code',
    'COUNTRY': 'country',
    'TERRITORY': 'territory',
    'CONTACTLASTNAME': 'contact_last_name',
    'CONTACTFIRSTNAME': 'contact_first_name',
    'DEALSIZE': 'deal_size'
})
```

```python
df.head()
```

```python
df['sales'].describe()
```

```python
sales_per_city = df.groupby('city')['sales'].sum()
print(sales_per_city)
```

```python
# standerizing the data

from sklearn.preprocessing import StandardScaler

# Assuming 'sales' is the column you want to standardize
scaler = StandardScaler()
df['sales_standardized'] = scaler.fit_transform(df[['sales']])

df.head()
```

```python
#  Data Type Validation
print("\nData Types:\n", df.dtypes)
```

```python
df['MONTH'] = df['order_date'].dt.month
monthly_sales = df.groupby('MONTH')['sales'].sum()
print(monthly_sales)
```
